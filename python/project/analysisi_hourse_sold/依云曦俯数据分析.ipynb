# 2、分析9栋出售情况

```python
import pandas as pd
building_9_data = pd.read_excel('./依云曦府.xlsx', index=False, sheet_name='9栋')


building_9_unsold_data = building_9_data[building_9_data['销售状态'] == '可售']
building_9_unsold_data
```

```python
building_9_unsold_data.建筑面积.unique()
```

```python
building_9_data_100 = building_9_data[building_9_data.建筑面积 == 100.8].shape[0]
building_9_data_124 = building_9_data[building_9_data.建筑面积 == 124.14].shape[0]
building_9_data_142 = building_9_data[building_9_data.建筑面积 == 142.57].shape[0]

building_9_unsold_data_100 = building_9_unsold_data[building_9_unsold_data.建筑面积 == 100.8].shape[0]
building_9_unsold_data_124 = building_9_unsold_data[building_9_unsold_data.建筑面积 == 124.14].shape[0]
building_9_unsold_data_142 = building_9_unsold_data[building_9_unsold_data.建筑面积 == 142.57].shape[0]
print('9栋总共{}套房间，其中{}套100面积房间，{}套124面积房间，{}套142面积房间'.format(building_9_data.shape[0],
                                                            building_9_data_100, building_9_data_124, building_9_data_142))
print('9栋剩余{}套房间，其中{}套100面积房间，{}套124面积房间，{}套142面积房间'.format(building_9_unsold_data.shape[0],
                                                           building_9_unsold_data_100, building_9_unsold_data_124, building_9_unsold_data_142))
```

```python
'''
输入dataframe数据
返回dataframe数据
'''
import numpy as np

def analysis_house_sold(data):
    cnt, sold_cnt, unsold_cnt = [], [], []
    
    unsold_data = data[data.销售状态 == '可售']
    areas = np.sort(unsold_data.建筑面积.unique())
    #print(areas)
    
    cnt.append(data.shape[0])
    unsold_cnt.append(unsold_data.shape[0])
    
    #计算出售情况
    for area in areas:
        cnt.append(data[data.建筑面积 == area].shape[0])
        unsold_cnt.append(unsold_data[unsold_data.建筑面积 == area].shape[0])
    return cnt, unsold_cnt

cnt, unsold_cnt = analysis_house_sold(data)
```
